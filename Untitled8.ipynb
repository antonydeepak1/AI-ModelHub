```python
from transformers import AutoModelForCausalLM, AutoProcessor, BitsAndBytesConfig
from google.colab import files
from PIL import Image
import cv2
import torch
import os
```

```python
login()
# STEP 3: Load the model with 4-bit quantization
model_id = "DAMO-NLP-SG/VideoLLaMA3-7B"
cache_dir = "/content/sample_data/models_VideoLLaMA3-7B"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.float16
)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True,
    torch_dtype=torch.bfloat16,
    attn_implementation="flash_attention_2",
    cache_dir=cache_dir
)

processor = AutoProcessor.from_pretrained(
    model_id,
    trust_remote_code=True,
    cache_dir=cache_dir
)
print("📁 Upload an image file")
image_upload = files.upload()

print("📁 Upload a video file")
video_upload = files.upload()

# Get file names from uploads
image_path = next(iter(image_upload))
video_path = next(iter(video_upload))
# STEP 5: Process uploaded image and video
def process_image(image_path):
    image = Image.open(image_path).convert("RGB")
    return image

def process_video(video_path, max_frames=128, fps=1):
    video_capture = cv2.VideoCapture(video_path)
    frames = []
    frame_rate = video_capture.get(cv2.CAP_PROP_FPS)
    total_frames = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_interval = int(frame_rate / fps) if frame_rate > 0 else 1
    for frame_idx in range(0, total_frames, frame_interval):
        ret, frame = video_capture.read()
        if not ret:
            break
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frames.append(Image.fromarray(frame))
        if len(frames) >= max_frames:
            break
    video_capture.release()
    return frames
    # STEP 6: Ask user for a prompt
prompt = input("💬 Enter your prompt (e.g., 'Describe the image and summarize the video.'):\n")
# STEP 7: Prepare inputs for the model
image = process_image(image_path)
video_frames = process_video(video_path)

conversation = [
    {"role": "system", "content": "You are a helpful assistant."},
    {
        "role": "user",
        "content": [
            {"type": "image", "image": image},
            {"type": "video", "video": {"video_path": video_path, "fps": 1, "max_frames": 128}},
            {"type": "text", "text": prompt}
        ]
    }
]
# STEP 8: Run the model and get the output
inputs = processor(conversation=conversation, return_tensors="pt")
inputs = {k: v.to(model.device) if isinstance(v, torch.Tensor) else v for k, v in inputs.items()}

output_ids = model.generate(**inputs, max_new_tokens=256)
response = processor.batch_decode(output_ids, skip_special_tokens=True)[0].strip()

print("\n✅ Model Output:\n")
print(response)

```
